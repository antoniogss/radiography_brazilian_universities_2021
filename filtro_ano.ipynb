```python
%%sql

/*

≡ CHART TYPE ≡

◔ Pie

≡ CHART SETTINGS ≡

★ Variable name: SG_IES
► Variable type: Text
► Filter widget label: SIGLA
► How should users filter on this variable?: Input box
► Required?: Enabled
► Default filter widget value: UNIFESP

★ Variable name: ANO
► Variable type: Number
► Filter widget label: ANO
► How should users filter on this variable?: Input box
► Required?: Enabled
► Default filter widget value: 2023

≡ DASHBOARD SETTINGS ≡

► Title: Número de Cursos por Grau de Especialização
► Description: Dados anuais filtrados por sigla da instituição de ensino e ano, coletados
pela Coordenação de Aperfeiçoamento de Pessoal de Nível Superior (CAPES) -
Baseado em consulta SQL com duplo critério de filtragem.
► Hide this card if there are no results: Enabled

*/

WITH ContagemCategoria AS (
    SELECT
        CASE
            WHEN NM_GRAU_CURSO = 'MESTRADO PROFISSIONAL' THEN 'MESTRADO PROFISSIONAL'
            WHEN NM_GRAU_CURSO = 'MESTRADO' THEN 'MESTRADO ACADÊMICO'
            WHEN NM_GRAU_CURSO = 'DOUTORADO' THEN 'DOUTORADO ACADÊMICO'
            WHEN NM_GRAU_CURSO = 'DOUTORADO PROFISSIONAL' THEN 'DOUTORADO PROFISSIONAL'
            ELSE 'OUTRO'
        END AS categoria,
        COUNT(DISTINCT CD_PROGRAMA_IES) AS quantidade
    FROM
        capes_21_22_23
    WHERE
        SG_IES = {{SG_IES}}
        AND ANO = {{ANO}}
    GROUP BY
        categoria
)
SELECT
    categoria,
    quantidade
FROM
    ContagemCategoria;
```
