```python
%%sql

/*

 CHART TYPE 

  Table

 CHART SETTINGS 

 Variable name: NO_IES
 Variable type: Text
 Filter widget label: NOME
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: XXXX

 Variable name: NOME_DO_MUNICIPIO
 Variable type: Text
 Filter widget label: NOME DO MUNICPIO
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: XXXX

 Variable name: ANO
 Variable type: Number
 Filter widget label: ANO
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: 2023

 DASHBOARD SETTINGS 

 Title: Listagem de Cursos Oferecidos
 Description: Censo da Educao Superior: dados anuais filtrados por nome da
instituio de ensino, nome do municpio de oferta de cursos e ano, coletados pelo
Instituto Nacional de Estudos e Pesquisas Educacionais Ansio Teixeira (INEP),
incluindo tambm dados do Exame Nacional de Desempenho dos Estudantes (ENADE) -
Baseado em consulta SQL com triplo critrio de filtragem.
 Hide this card if there are no results: Enabled

*/

SELECT
    UPPER(COALESCE(ces.NO_CURSO, 'NO INDEXADO')) AS "NOME DO CURSO",
    CASE
        WHEN ces.TP_GRAU_ACADEMICO = '1' THEN 'BACHARELADO'
        WHEN ces.TP_GRAU_ACADEMICO = '2' THEN 'LICENCIATURA'
        WHEN ces.TP_GRAU_ACADEMICO = '3' THEN 'TECNOLGICO'
        ELSE 'NO INDEXADO'
    END AS "GRAU ACADMICO",
    CASE
        WHEN ces.TP_MODALIDADE_ENSINO = '1' THEN 'PRESENCIAL'
        WHEN ces.TP_MODALIDADE_ENSINO = '2' THEN 'CURSO A DISTNCIA'
        ELSE 'NO INDEXADO'
    END AS "MODALIDADE DE ENSINO",
    UPPER(COALESCE(ces.NO_CINE_AREA_GERAL, 'NO INDEXADO')) AS "GRANDE REA DO CONHECIMENTO",
    UPPER(COALESCE(ces.NO_MUNICIPIO, 'NO INDEXADO')) AS "NOME DO MUNICPIO",
    UPPER(COALESCE(ces.NO_UF, 'NO INDEXADO')) AS "ESTADO",
    COALESCE(CASE WHEN enade.CONCEITO_ENADE_CONTINUO IS NULL THEN 'NO AVALIADO' ELSE ROUND(enade.CONCEITO_ENADE_CONTINUO)::TEXT END, 'NO AVALIADO') AS "CONCEITO ENADE CONTNUO",
    COALESCE(UPPER(enade.ANO::TEXT), 'NO REALIZADO') AS "EDIO DO ENADE"
FROM
    ces_21_22_23 ces
LEFT JOIN
    enade_21_22_23 enade
ON
    ces.CO_CURSO = enade.CO_CURSO
WHERE
    ces.NO_IES = {{NO_IES}}
    AND ces.ANO = {{ANO}}
    AND (UPPER(COALESCE(ces.NO_MUNICIPIO, 'NO INDEXADO')) = {{NOME_DO_MUNICIPIO}})
ORDER BY "NOME DO CURSO";
```

```python
%%sql

/*

 CHART TYPE 

 Pie

 CHART SETTINGS 

 Variable name: SG_IES
 Variable type: Text
 Filter widget label: SIGLA
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: UNIFESP

 Variable name: ANO
 Variable type: Number
 Filter widget label: ANO
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: 2023

 DASHBOARD SETTINGS 

 Title: Nmero de Cursos por Grau de Especializao
 Description: Dados anuais filtrados por sigla da instituio de ensino e ano, coletados
pela Coordenao de Aperfeioamento de Pessoal de Nvel Superior (CAPES) -
Baseado em consulta SQL com duplo critrio de filtragem.
 Hide this card if there are no results: Enabled

*/

WITH ContagemCategoria AS (
    SELECT
        CASE
            WHEN NM_GRAU_CURSO = 'MESTRADO PROFISSIONAL' THEN 'MESTRADO PROFISSIONAL'
            WHEN NM_GRAU_CURSO = 'MESTRADO' THEN 'MESTRADO ACADMICO'
            WHEN NM_GRAU_CURSO = 'DOUTORADO' THEN 'DOUTORADO ACADMICO'
            WHEN NM_GRAU_CURSO = 'DOUTORADO PROFISSIONAL' THEN 'DOUTORADO PROFISSIONAL'
            ELSE 'OUTRO'
        END AS categoria,
        COUNT(DISTINCT CD_PROGRAMA_IES) AS quantidade
    FROM
        capes_21_22_23
    WHERE
        SG_IES = {{SG_IES}}
        AND ANO = {{ANO}}
    GROUP BY
        categoria
)
SELECT
    categoria,
    quantidade
FROM
    ContagemCategoria;
```

```python
%%sql

/*

 CHART TYPE 

  Table

 CHART SETTINGS 

 Variable name: NO_IES
 Variable type: Text
 Filter widget label: NOME
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: XXXX

 Variable name: NOME_DO_MUNICIPIO
 Variable type: Text
 Filter widget label: NOME DO MUNICPIO
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: XXXX

 Variable name: ANO
 Variable type: Number
 Filter widget label: ANO
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: 2023

 DASHBOARD SETTINGS 

 Title: Listagem de Cursos Oferecidos
 Description: Dados anuais filtrados por nome da instituio de ensino, nome
do municpio de oferta de cursos e ano,  coletados pela Coordenao de Aperfeioamento
de Pessoal de Nvel Superior (CAPES) - Baseado em consulta SQL com triplo critrio de filtragem.
 Hide this card if there are no results: Enabled

*/

WITH CursoData AS (
    SELECT
        CASE
            WHEN NM_GRAU_CURSO = 'DOUTORADO' THEN 'DOUTORADO ACADMICO'
            WHEN NM_GRAU_CURSO = 'MESTRADO' THEN 'MESTRADO ACADMICO'
            ELSE COALESCE(UPPER(NM_GRAU_CURSO), 'NO INDEXADO')
        END AS "GRAU ACADMICO",
        UPPER(COALESCE(NM_CURSO, 'NO INDEXADO')) AS "NOME DO CURSO",
        UPPER(COALESCE(NM_GRANDE_AREA_CONHECIMENTO, 'NO INDEXADO')) AS "GRANDE REA DO CONHECIMENTO",
        UPPER(COALESCE(NM_MUNICIPIO_PROGRAMA_IES, 'NO INDEXADO')) AS "NOME DO MUNICPIO",
        CASE
            WHEN SG_UF_IES = 'AC' THEN 'ACRE'
            WHEN SG_UF_IES = 'AL' THEN 'ALAGOAS'
            WHEN SG_UF_IES = 'AP' THEN 'AMAP'
            WHEN SG_UF_IES = 'AM' THEN 'AMAZONAS'
            WHEN SG_UF_IES = 'BA' THEN 'BAHIA'
            WHEN SG_UF_IES = 'CE' THEN 'CEAR'
            WHEN SG_UF_IES = 'DF' THEN 'DISTRITO FEDERAL'
            WHEN SG_UF_IES = 'ES' THEN 'ESPRITO SANTO'
            WHEN SG_UF_IES = 'GO' THEN 'GOIS'
            WHEN SG_UF_IES = 'MA' THEN 'MARANHO'
            WHEN SG_UF_IES = 'MT' THEN 'MATO GROSSO'
            WHEN SG_UF_IES = 'MS' THEN 'MATO GROSSO DO SUL'
            WHEN SG_UF_IES = 'MG' THEN 'MINAS GERAIS'
            WHEN SG_UF_IES = 'PA' THEN 'PAR'
            WHEN SG_UF_IES = 'PB' THEN 'PARABA'
            WHEN SG_UF_IES = 'PR' THEN 'PARAN'
            WHEN SG_UF_IES = 'PE' THEN 'PERNAMBUCO'
            WHEN SG_UF_IES = 'PI' THEN 'PIAU'
            WHEN SG_UF_IES = 'RJ' THEN 'RIO DE JANEIRO'
            WHEN SG_UF_IES = 'RN' THEN 'RIO GRANDE DO NORTE'
            WHEN SG_UF_IES = 'RS' THEN 'RIO GRANDE DO SUL'
            WHEN SG_UF_IES = 'RO' THEN 'RONDNIA'
            WHEN SG_UF_IES = 'RR' THEN 'RORAIMA'
            WHEN SG_UF_IES = 'SC' THEN 'SANTA CATARINA'
            WHEN SG_UF_IES = 'SP' THEN 'SO PAULO'
            WHEN SG_UF_IES = 'SE' THEN 'SERGIPE'
            WHEN SG_UF_IES = 'TO' THEN 'TOCANTINS'
            ELSE COALESCE(SG_UF_IES, 'NO INDEXADO')
        END AS "ESTADO",
        CD_CONCEITO_CURSO AS "CONCEITO CAPES",
        TO_CHAR(AN_BASE, '9999') AS "EDIO"
    FROM
        capes_21_22_23 capes
    WHERE
        capes.NO_IES = {{NO_IES}}
        AND capes.ANO = {{ANO}}
        AND (UPPER(COALESCE(capes.NM_MUNICIPIO_PROGRAMA_IES, 'NO INDEXADO')) = {{NM_MUNICIPIO_PROGRAMA_IES}})
)
SELECT "NOME DO CURSO", "GRAU ACADMICO", "GRANDE REA DO CONHECIMENTO", "NOME DO MUNICPIO", "ESTADO", "CONCEITO CAPES", "EDIO"
FROM CursoData
ORDER BY "NOME DO CURSO";
```

* **{{NOME_DO_MUNICIPIO}}**  o filtro do Metabase que representa o nome do municpio onde a IES oferece o curso;
* **{{ANO}}**  o novo filtro que voc adicionar no Metabase para filtrar por ano, configurar como valor obrigatrio sempre o ano mais recente.

```python
%%sql

/*

 CHART TYPE 

  Table

 CHART SETTINGS 

 Variable name: SG_IES
 Variable type: Text
 Filter widget label: SIGLA
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: UNIFESP

 Variable name: NOME_DO_MUNICIPIO
 Variable type: Text
 Filter widget label: NOME DO MUNICPIO
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: XXXX

 Variable name: ANO
 Variable type: Number
 Filter widget label: ANO
 How should users filter on this variable?: Input box
 Required?: Enabled
 Default filter widget value: 2023

 DASHBOARD SETTINGS 

 Title: Listagem de Cursos Oferecidos
 Description: Dados anuais filtrados por nome da instituio de ensino, nome
do municpio de oferta de cursos e ano,  coletados pela Coordenao de Aperfeioamento
de Pessoal de Nvel Superior (CAPES) - Baseado em consulta SQL com triplo critrio de filtragem.
 Hide this card if there are no results: Enabled

*/

WITH CursoData AS (
    SELECT
        CASE
            WHEN NM_GRAU_CURSO = 'DOUTORADO' THEN 'DOUTORADO ACADMICO'
            WHEN NM_GRAU_CURSO = 'MESTRADO' THEN 'MESTRADO ACADMICO'
            ELSE COALESCE(UPPER(NM_GRAU_CURSO), 'NO INDEXADO')
        END AS "GRAU ACADMICO",
        UPPER(COALESCE(NM_CURSO, 'NO INDEXADO')) AS "NOME DO CURSO",
        UPPER(COALESCE(NM_GRANDE_AREA_CONHECIMENTO, 'NO INDEXADO')) AS "GRANDE REA DO CONHECIMENTO",
        UPPER(COALESCE(NM_MUNICIPIO_PROGRAMA_IES, 'NO INDEXADO')) AS "NOME DO MUNICPIO",
        CASE
            WHEN SG_UF_IES = 'AC' THEN 'ACRE'
            WHEN SG_UF_IES = 'AL' THEN 'ALAGOAS'
            WHEN SG_UF_IES = 'AP' THEN 'AMAP'
            WHEN SG_UF_IES = 'AM' THEN 'AMAZONAS'
            WHEN SG_UF_IES = 'BA' THEN 'BAHIA'
            WHEN SG_UF_IES = 'CE' THEN 'CEAR'
            WHEN SG_UF_IES = 'DF' THEN 'DISTRITO FEDERAL'
            WHEN SG_UF_IES = 'ES' THEN 'ESPRITO SANTO'
            WHEN SG_UF_IES = 'GO' THEN 'GOIS'
            WHEN SG_UF_IES = 'MA' THEN 'MARANHO'
            WHEN SG_UF_IES = 'MT' THEN 'MATO GROSSO'
            WHEN SG_UF_IES = 'MS' THEN 'MATO GROSSO DO SUL'
            WHEN SG_UF_IES = 'MG' THEN 'MINAS GERAIS'
            WHEN SG_UF_IES = 'PA' THEN 'PAR'
            WHEN SG_UF_IES = 'PB' THEN 'PARABA'
            WHEN SG_UF_IES = 'PR' THEN 'PARAN'
            WHEN SG_UF_IES = 'PE' THEN 'PERNAMBUCO'
            WHEN SG_UF_IES = 'PI' THEN 'PIAU'
            WHEN SG_UF_IES = 'RJ' THEN 'RIO DE JANEIRO'
            WHEN SG_UF_IES = 'RN' THEN 'RIO GRANDE DO NORTE'
            WHEN SG_UF_IES = 'RS' THEN 'RIO GRANDE DO SUL'
            WHEN SG_UF_IES = 'RO' THEN 'RONDNIA'
            WHEN SG_UF_IES = 'RR' THEN 'RORAIMA'
            WHEN SG_UF_IES = 'SC' THEN 'SANTA CATARINA'
            WHEN SG_UF_IES = 'SP' THEN 'SO PAULO'
            WHEN SG_UF_IES = 'SE' THEN 'SERGIPE'
            WHEN SG_UF_IES = 'TO' THEN 'TOCANTINS'
            ELSE COALESCE(SG_UF_IES, 'NO INDEXADO')
        END AS "ESTADO",
        CD_CONCEITO_CURSO AS "CONCEITO CAPES",
        TO_CHAR(AN_BASE, '9999') AS "EDIO"
    FROM
        capes_21_22_23 capes
    WHERE
        capes.SG_IES = {{SG_IES}}
        AND capes.ANO = {{ANO}}
        AND (UPPER(COALESCE(capes.NM_MUNICIPIO_PROGRAMA_IES, 'NO INDEXADO')) = {{NM_MUNICIPIO_PROGRAMA_IES}})
)
SELECT "NOME DO CURSO", "GRAU ACADMICO", "GRANDE REA DO CONHECIMENTO", "NOME DO MUNICPIO", "ESTADO", "CONCEITO CAPES", "EDIO"
FROM CursoData
ORDER BY "NOME DO CURSO";
```
